# 熵，信息熵，香农熵

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['font.sans-serif']=['STHeiti','sans-serif']

```

## 要得出熵，首先要得出信息量，信息量度量的是一个具体事件发生了所带来的信息

```python
# p_x 为概率，那么则是0~1之间的数
p_x = np.arange(0.,1.0,0.01)[1:]

# 信息？信息量？
l_x_ = np.log2(p_x) # 信息量需要为正数，所以应该再加个负号
l_x = -l_x_
plt.figure()
l1, = plt.plot(p_x,l_x_,'r-', label='')
l2, = plt.plot(p_x, l_x, 'g-', label='信息量')
plt.legend([l1,l2],['','信息量'])
plt.grid(True)
plt.xlabel('事件发生的概率',fontsize='large')
plt.ylabel('信息量的大小',fontsize='large')
plt.show()
```

信息量与事件发生的概率有关，越小概率的事情发生了产生的信息量越大，越大概率的事情发生了产生的信息量越小  

## 熵则是在结果出来之前对可能产生的信息量的期望——考虑该随机变量的所有可能取值，即所有可能发生事件所带来的信息量的期望

信息熵还可以作为一个系统复杂程度的度量，如果系统越复杂，出现不同情况的种类越多，那么他的信息熵是比较大的
